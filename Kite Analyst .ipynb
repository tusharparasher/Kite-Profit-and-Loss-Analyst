```python
import pandas as pd
import numpy as np
import os
import re
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("orders (1).csv")
```

```python
df
```

```python
df = df[df["Status"]!="REJECTED"]
df
```

```python
df = df[df["Product"]!="CNC"]
df
```

```python
df["Qty."] =  df["Qty."].apply(lambda x: x.split("/")[0])
```

```python
df["Qty."] = df["Qty."].astype(int)
df
```

```python
df.info()
```

```python
df["Total"] = df["Qty."]*df["Avg. price"]
```

```python
df
```

```python
df["GrandPnL"] = np.where((df["Type"]=="BUY"),-1*df["Total"],df["Total"])
```

```python
df
```

```python
df["Charge"] = np.minimum(20,(df["Total"]*0.03/100))
```

```python
df
```

```python
df["ETC"] = df["Total"]*0.003/100
```

```python
df
```

```python
df["GST"] = df[["Charge","ETC"]].sum(axis="columns")*18/100
df
```

```python
df["STT"] = np.where(df["Type"]=="BUY",0,df["Total"]*0.025/100)
df
```

```python
df["Stamp duty"] = np.where(df["Type"]=="BUY",df["Total"]*0.003/100,0)
df
```

```python
df["Total Charges"] = df[["Charge","ETC","GST","STT","Stamp duty"]].sum(axis="columns")
df
```

```python
df.to_excel("kite calculation2.xlsx",index=False)
```

```python
df_summary = pd.DataFrame(df.groupby(['Instrument'])['GrandPnL','Total Charges'].sum()).reset_index()
```

```python
df_summary
```

```python
df_summary['Net PnL']=df_summary['GrandPnL']-df_summary['Total Charges']
```

```python
df_summary['% Charge']=(df_summary['Total Charges']/df_summary['GrandPnL'])*100
df_summary
```

```python
df_total = pd.pivot_table(pd.DataFrame(df_summary.sum(axis=0)).reset_index(),columns=['index'],values=[0],aggfunc=lambda x: ' '.join(str(v) for v in x))
df_total['Instrument']='Total'
```

```python
df_total
```

```python
df_final = pd.concat([df_summary,df_total],axis='rows')
```

```python
df_final
```

```python
list_columns = ['GrandPnL','Total Charges','Net PnL','% Charge']
for i in list_columns:
    df_final[i]=df_final[i].astype(float)
    df_final[i]=df_final[i].round(2)
print(df_final.info())
```

```python
writer = pd.ExcelWriter('kite_pnl_calculated.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Data', index=False)
df_final.to_excel(writer, sheet_name='Summary', index=False)
writer.save()
df_final.iloc[-1,-1] = round((df_final.iloc[-1,2]/df_final.iloc[-1,1])*100,2)
```
